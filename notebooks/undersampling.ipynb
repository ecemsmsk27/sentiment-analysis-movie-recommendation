```python
import pandas as pd
```

```python
df = pd.read_csv("/content/etiketlenmis_83k - etiketlenmis_83k.csv")
```

```python
df.head()
```

```python
def is_quality_comment(comment):
    if not isinstance(comment, str):
        return False
    if len(comment.strip()) < 30:  # çok kısa yorumları çıkar
        return False
    if comment.lower().count("film") + comment.lower().count("oyuncu") < 1:
        return False  # çok alakasız yorumları çıkar
    if len(set(comment.lower().split())) < 4:
        return False  # hep aynı kelimeler varsa çıkar
    return True

df = df[df["comment"].apply(is_quality_comment)]
```

```python
# Her sınıftan kaliteli 7000 örnek al
target_labels = ["izlemeye_deger", "basyapit", "vasat", "zaman_kaybi", "beklentiyi_karsilamadi"]

balanced_df = pd.DataFrame()

for label in target_labels:
    subset = df[df["predicted_label"] == label]
    if len(subset) >= 7000:
        sample = subset.sample(n=7000, random_state=42)
    else:
        sample = subset  # elimizde yeterince yoksa tamamını al
    balanced_df = pd.concat([balanced_df, sample], ignore_index=True)

# Sonuç CSV
balanced_df.to_csv("balanced_fine_grained.csv", index=False)
print("Dengeli veri seti başarıyla oluşturuldu.")
```
